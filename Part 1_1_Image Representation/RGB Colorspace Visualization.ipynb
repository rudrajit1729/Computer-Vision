# RGB Colorspace

## Import Libraries

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

## Read Input Image

```python
# Read Input Image
image = mpimg.imread('images/wa_state_highway.jpg')

# Display Image
plt.imshow(image)
```

## RGB Colorspace 

Visualize the levels of each color channel. Take a closer look at the traffic signs. 

```python
# Isolate the RGB Channels
r = image[:, :, 0]
g = image[:, :, 1]
b = image[:, :, 2]

# Visualize the color channels individually
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (20, 10))
ax1.set_title('R Channel')
ax1.imshow(r, cmap = 'gray')
ax2.set_title('G Channel')
ax2.imshow(g, cmap = 'gray')
ax3.set_title('B Channel')
ax3.imshow(b, cmap = 'gray')
```
